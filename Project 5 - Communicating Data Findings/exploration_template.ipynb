A solution for this problem is that when exploring the variables to graph, we could explore the relationships of variables that don't have missing values, or few missing values.

```python
#Univariate Analysis


#Let's explore the different loan terms
terms = loan['Term'].unique()
terms_ordered_int = sorted(terms)

#I created a new array here with the strings of the ordered terms so that bar chart would look cleaner with 
#just those three bins
terms_ordered = ['12','36','60']

#Let's now create an array of size 3 to hold the counts that those terms appeared
terms_array = loan['Term']
terms_count = [0,0,0]
for i in range(len(loan)):
    if terms_array[i] == 12:
        terms_count[0] += 1
    elif terms_array[i] == 36:
        terms_count[1] += 1
    else:
        terms_count[2] += 1

# Graphing the individual counts of terms used in order to see which loan term length was the most popular
x = terms_ordered
y = terms_count
plt.bar(x,y)
plt.xlabel('Term lengths of loans (Months)')
plt.ylabel('Number of loans')
plt.title('Number of loans per term length')
```

After visualizing the number of loans per term length, we can see that a loan term length of 36 months was the most popular among this dataset. 

```python
#This time, let's investigate which income range is the most popular in this dataset
income_range = loan['IncomeRange'].unique()
#Here, we are hand ordering our income range into an array because of the mixture of number range but also strings
income_range_ordered = ['Not employed','$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+','Not displayed']

#Let's now create an array of size 8 to hold the counts that those income ranges appeared
income_array = loan['IncomeRange']
income_count = [0,0,0,0,0,0,0,0]
for i in range(len(loan)):
    if income_array[i] == 'Not employed':
        income_count[0] += 1
    elif income_array[i] == '$0':
        income_count[1] += 1
    elif income_array[i] == '$1-24,999':
        income_count[2] += 1
    elif income_array[i] == '$25,000-49,999':
        income_count[3] += 1
    elif income_array[i] == '$50,000-74,999':
        income_count[4] += 1
    elif income_array[i] == '$75,000-99,999':
        income_count[5] += 1
    elif income_array[i] == '$100,000+':
        income_count[6] += 1
    else:
        income_count[7] += 1

# Graphing the individual counts of terms used in order to see which income range is the most frequent
x = income_range_ordered
y = income_count
plt.bar(x,y)
plt.xlabel('Income range (Dollars $)')
plt.ylabel('Number of loans')
plt.xticks(rotation='vertical')
plt.title('Number of loans per income range')
```

From the visualization of the number of loans per income range, we can identify that the most number of loans came from the income range of $25,000-49,999. Looking at the shape of this plot, it seems that the number of loans per income range is a pretty normally distributed variable.

```python
#Next, investigate if there were more borrowers who are homeowners or not
borrowers = loan['IsBorrowerHomeowner'].unique()
#Hand wrote this just two have the two bins in the bar chart
borrowers_ordered = ['True','False']
#Let's now create an array of size 2 to hold the counts that borrowers are homeowners or not
borrower_array = loan['IsBorrowerHomeowner']
borrower_count = [0,0]
for i in range(len(loan)):
    if borrower_array[i] == True:
        borrower_count[0] += 1
    else:
        borrower_count[1] += 1

# Graphing the individual counts of terms used in order to see if there were more homeower borrowers or not
x = borrowers_ordered
y = borrower_count
plt.bar(x,y)
plt.xlabel('Borrower is a homeowner')
plt.ylabel('Number of loans')
plt.title('Number of loans based on if borrower is a homeowner or not')
```

From this visualization, we can see that the number of borrowers who were homeowners and who were not homeowners were indeed pretty close. However, looking at the figure carefully, we can see that there were a few more homeowners that were borrowers in this dataset.

```python
#Next, investigate if there were more borrowers who had verifiable incomes
income_verify = loan['IncomeVerifiable'].unique()
#Hand wrote this just two have the two bins in the bar chart
income_verify_ordered = ['True','False']
#Let's now create an array of size 2 to hold the counts that borrowers who had verifiable incomes
verify_array = loan['IncomeVerifiable']
verify_count = [0,0]
for i in range(len(loan)):
    if verify_array[i] == True:
        verify_count[0] += 1
    else:
        verify_count[1] += 1

# Graphing the individual counts of terms used in order to see if there were more borrowers who had verifiable incomes
x = income_verify_ordered
y = verify_count
plt.bar(x,y)
plt.xlabel('Borrower income verified')
plt.ylabel('Number of loans')
plt.title('Number of loans based on if borrower income is verified')
```

From this visualization, it is clear to say that a majority of borrowers incomes are verifiable. 

```python
#Bivariate Analysis


#Let's explore the relationship between the number of investors and whether or not the loan borrowers are 
#homeowners
sb.boxplot(x=loan['IsBorrowerHomeowner'], y=loan['Investors']).set_title("Number of inverstors based on whether or not the borrower is a homeowner")

```

From this visualization of the number of investors based on whether or not the borrower is a homeowner, we can see that the average number of investors, the higher bound/upper quartile, and the lower bound/lower quartile is higher when the borrower was a homeowner. We could also infer that due to the higher outliers when the borrowers were homeowners, that could have brought up the average, upper quartile, and lower quartile.

```python
#Let's see the relationship between the original loan amount and investor numbers
sb.lmplot(x='Investors',y='LoanOriginalAmount', data=loan)
plt.title("Relationship between the number of investors and the original loan amount")
plt.ylabel('LoanOriginalAmount (Dollars $)')
```

Looking at the visualization, it seems that the higher the original loan amount becomes, the higher the investor numbers. However, this point is not necessairly clear when we look at loan amounts over $25,000

```python
#Let's see the relationship between income range and original loan amount
sb.boxplot(x=sorted(loan['IncomeRange']), y=loan['LoanOriginalAmount']).set_title("Original Loan Amounts vs Income Range")
plt.xlabel('Income range (Dollars $)')
plt.ylabel('LoanOriginalAmount (Dollars $)')
plt.xticks(rotation='vertical')
```

After reviewing the visualization, most of the income ranges seem to have very similar average original loan amounts. However, the one that makes a pretty significant drop in the average original loan amount is when the income range is $0. This is somewhat expected, as when a person's income range is $0, then they should also not be getting higher original loan amounts.

```python
#Another question that was explored is whether or not longer terms leads to higher original loan amounts.
sb.boxplot(x=sorted(loan['Term']), y=loan['LoanOriginalAmount']).set_title("Original Loan Amounts vs Loan Term Length")
plt.xlabel('Term (Months)')
plt.ylabel('LoanOriginalAmount (Dollars $)')
```

Analyzing the visualization above, it seems like when the term is shortest, at 12 months, the average original loan amount was actually higher. However, the surprising part was that all the upper quartiles for each term length ended up being pretty similar to each other. 

```python
#Let's see the relationship between the original loan amount and monthly loan payments
sb.lmplot(x='MonthlyLoanPayment',y='LoanOriginalAmount', data=loan)
plt.title("Relationship between the original loan amount and monthly loan payments")
plt.xlabel('MonthlyLoanPayment (Dollars $)')
plt.ylabel('LoanOriginalAmount (Dollars $)')
```

Reviewing the visualization, especially the regression line, we can see that there is a fairly positive correlation between the montly loan payments and the original loan amounts. It seems like the higher the monthly loan payments are, the higher the original loan amounts were, which is something that I assumed when I first decided to explore this relationship.

```python
#Multivariate Analysis


#Let's explore the different income ranges of the borrowers
income_range = loan['IncomeRange'].unique()
#Here, we are hand ordering our income range into an array because of the mixture of number range but also strings
income_range_ordered = ['Not employed','$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+','Not displayed']

#Exploring the relationship between the number of investors and the monthly payments, then adding on top the 
#variable of the income range of the borrowers
sb.lmplot(x='MonthlyLoanPayment', y='Investors', data=loan, hue='IncomeRange', legend=False)
plt.legend(title='Income Range', loc='upper left', labels=['Not employed','$0','$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999','$100,000+','Not displayed'])
plt.title("Ratio of investors per monthly loan payment")
```

Looking at this visualization, we can see the linear regression lines for each income range. The limitations of this visualization is the fact that because there are so many data points, all of the individual points are very hard to see, as they are all clumped up. However, from analyzing the regression lines, it seems like there are typically the highest ratio of investors per monthly loan payment when the income range is $75,000-99,000.
