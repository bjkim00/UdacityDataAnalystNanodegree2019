## Investigation Overview

There were three main investigations made with this dataset. For the univariate analysis, I investigated which 
loan term was the most popular among this dataset. For the bivariate analysis, I explored the relationship between whether or not the number of investors increased or decreased based on if the borrower was a homeowner or not. Finally, I investigated the relationship of the number of investors in regards to the monthly loan payment amount by income ranges.

## Dataset Overview

The dataset gave access to many different aspects of loans. These included things like loan term length, montly loan payment amount, number of investors, loan listing key, etc. One problem that arose when trying to clean the data was that when I simply tried to get rid of any rows that had any null values, then in total, all of the rows in the dataset got removed. Thus, I comprised with this problem and made sure to base all my explorations of the dataset with rows/variables that didn't have any missing values. Regarding duplicated values, there were no duplicated rows within the dataset, so I did not have to worry about that when cleaning the data.

```python
#Univariate Analysis


#Let's explore the different loan terms
terms = loan['Term'].unique()
terms_ordered_int = sorted(terms)

#I created a new array here with the strings of the ordered terms so that bar chart would look cleaner with 
#just those three bins
terms_ordered = ['12','36','60']

#Let's now create an array of size 3 to hold the counts that those terms appeared
terms_array = loan['Term']
terms_count = [0,0,0]
for i in range(len(loan)):
    if terms_array[i] == 12:
        terms_count[0] += 1
    elif terms_array[i] == 36:
        terms_count[1] += 1
    else:
        terms_count[2] += 1

# Graphing the individual counts of terms used in order to see which loan term length was the most popular
x = terms_ordered
y = terms_count
plt.bar(x,y)
plt.xlabel('Term lengths of loans (Months)')
plt.ylabel('Number of loans')
plt.title('Number of loans per term length')
```
