After cleaning the data, I realized that my dataset had gotten a lot smaller. The latest date in the cleaned dataset is now August 2016, and the most recent date is September 2017.

Create a new dataframe to hold the totals for each different month.

```python
new_df = pd.DataFrame(gun['month'].unique(), columns=['Month'])
new_df.head()
```

Go through the csv file to total up the values under the 'total' column for each month.

After all the totals have been put in a series, add to the new dataframe with only months. This will be used to visualize the trend of gun purchases.

```python
gun_copy = gun
total_per_month = []
traverse_month = new_df['Month'].unique()
for i in traverse_month:
    temp = gun_copy[gun_copy['month'] == i]
    temp_totals = temp['totals']
    total = 0
    for j in temp_totals:
        total += j
    total_per_month.append(total)

new_df['Totals'] = total_per_month
```

Double check to make sure the totals have been added to the new dataframe correctly

```python
new_df
```

Now create a visualization to see how the gun purchase trends have been starting from August 2016 to September 2017 

```python
x = new_df['Month']
y = new_df['Totals']
plt.plot(x, y, color='g')
plt.xticks(rotation='vertical')
plt.title('Gun purchases between August 2016 and September 2019')
plt.xlabel('Month')
plt.ylabel('Number of guns purchased')
plt.show()
```

The visualization shows the gun purchases the past year between August 2016 and September 2017. Over that year, it seems that the gun purchases has gone up and down, hitting a maximum purchase of near 2,800,000 guns, and hitting a minimum purhcase of a little under 1,800,000 guns.

Let's see if there is a correlation between the number of permits and number of guns purchased. I will do this by creating 4 different regions to describe the US. Northeast, Midwest, South, and West. After separating the categories, the total number of permits and total number of guns purchased will be put into their respective regions. Then, we will create one bar graph showing the regions respective ratio of permit counts to total guns purchased.

```python
#Just creating a copy of the cleaned gun dataframe
gun_clean = gun

#Create a new dataframe for the regions
regions_df = pd.DataFrame()
regions = ['Northeast','Midwest', 'South', 'West']
regions_df['Regions'] = regions

#At this point, the four regions have been created, now it is time to create lists for which states go in which regions
northeast = ['Connecticut', 'Maine', 'Massachusetts', 'New Hampshire', 'Rhode Island', 'Vermont', 'New Jersey', 'New York', 'Pennsylvania']
midwest = ['Illinois', 'Indiana', 'Michigan', 'Ohio', 'Wisconsin', 'Iowa', 'Kansas', 'Minnesota', 'Missouri', 'Nebraska', 'North Dakota', 'South Dakota']
south = ['Delaware', 'Florida', 'Georgia', 'Maryland', 'North Carolina', 'South Carolina', 'Virginia', 'District of Columbia', 'West Virginia', 'Alabama', 'Kentucky', 'Mississippi', 'Tennessee', 'Arkansas', 'Louisiana', 'Oklahoma', 'Texas']
west = ['Arizona', 'Colorado', 'Idaho', 'Montana', 'Nevada', 'New Mexico', 'Utah', 'Wyoming', 'Alaska', 'California', 'Hawaii', 'Oregon', 'Washington']

#Now, we will create some lists to hold the state names, their relative permit numbers, and the total guns purchased by
#so that we can add them to the new dataframe
states = gun_clean['state']
permit = gun['permit']
total = gun['totals']
permit_per_region = [0,0,0,0]
total_guns_per_region = [0,0,0,0]

#Now traverse through the lists and add them together to their respective regions
for i in range(len(gun_clean)):
    if states[i] in northeast:
        permit_per_region[0] += permit[i]
        total_guns_per_region[0] += total[i]
    elif states[i] in midwest:
        permit_per_region[1] += permit[i]
        total_guns_per_region[1] += total[i]
        continue
    elif states[i] in south:
        permit_per_region[2] += permit[i]
        total_guns_per_region[2] += total[i]
    else:
        permit_per_region[3] += permit[i]
        total_guns_per_region[3] += total[i]

#Now that all of the regions have had their permit counts and total gun purchases totaled, we should be able to add them
#to the dataframe
regions_df['Total permit count'] = permit_per_region
regions_df['Total gun purchases'] = total_guns_per_region
regions_df
```

This is the total permit count and total gun purchases from the cleaned gun data, thus only from August 2016 to September 2017

Now that we have all the permit counts, we can visualize with bar graphs to see which region had the highest permit to gun ratio

```python
#Lets now calculate the amount of gun purchases per permit count
gun_purchases_per_permit = []
for i in range(len(regions_df)):
    ratio = total_guns_per_region[i]/permit_per_region[i]
    gun_purchases_per_permit.append(ratio)

#Now add this list to the dataframe
regions_df['Proportion of gun purchases per permit count'] = gun_purchases_per_permit
regions_df
```

The main reason that I ended up looking at the proportion rather than the just the total gun purchase numbers is because there is such a big difference between the maximum total gun purchases and minimum total gun purchases. Thus it would be harder to create a good visualization.

```python
x = regions_df['Regions']
y = regions_df['Proportion of gun purchases per permit count']
plt.bar(x, y)
plt.xlabel('Regions of US')
plt.ylabel('Proportion of gun purchases per permit count')
plt.title('Proportion of gun purchases per permit count by regions of US')
plt.show()
```

We can see from the visualization that the Northeast region of the United States had the highest proportion of gun purchases per permit count within the time period of August 2016 to September 2017.
